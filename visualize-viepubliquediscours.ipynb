```python
# load embedddings
import joblib

dterm_hiddenstate = joblib.load("/Users/yukatherin/Downloads/vie-publique-discours/embeddings/communautarisme.joblib")
print(len(dterm_hiddenstate))

term_text_sequences = joblib.load("/Users/yukatherin/Downloads/vie-publique-discours/embeddings/communautarisme2_term_text_sequences.joblib")
print(len(term_text_sequences))
```

```python
# fit tsne
from sklearn.manifold import TSNE

tsne = TSNE(random_state=1, n_iter=1000, metric="cosine")

embs = tsne.fit_transform(dterm_hiddenstate)
```

```python
import matplotlib.pyplot as plt

FS = (15, 10)
figx = plt.figure(figsize=FS)
plt.scatter(embs[:, 0], embs[:, 1], alpha=.1, linewidths=1, color='blue', marker='o', s=100)
plt.title("contextualized word embeddings of 265 different uses of the word `communautarisme`— (t-sne)", size=16)

# annotate groups
plt.annotate("group 1", (-5, 10), size=12)
plt.annotate("group 2", (-15, -8), size=12)
plt.annotate("group 3", (8, -8), size=12)
```

```python
import numpy as np

# group 1 
g1 = np.where(embs[:, 1] > 5)[0]
print(g1)

# group 2
g2 = np.where(np.logical_and(embs[:, 1] < 0, embs[:, 0] < 0))[0]
print(g2)

# group 3
g3 = np.where(np.logical_and(embs[:, 1] < 0, embs[:, 0] > 0))[0]
print(g3)

```

```python
# group 1 
for i in g1:
    print(i, term_text_sequences[i])
```

```python
# group 2
for i in g2:
    print(i, term_text_sequences[i])
```

```python
# group 3
for i in g3:
    print(i, term_text_sequences[i])
```
